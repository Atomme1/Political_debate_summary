```python
os.chdir("wav/")

pattern_one_word = ["s001","s002","s003","s005","s006","s007","s014","s015","s023","s024"]

for pattern in pattern_one_word:
    wav_files = [f for f in os.listdir() if pattern in f]
    print(wav_files)
    for wav in wav_files:
        new_path = '../wav_one_word/' + wav
        shutil.move(wav, new_path)

```
